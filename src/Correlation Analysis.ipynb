#Finding the Pearson and Spearman correlation coefficients between tweet data and price

```python
import pandas as pd


file_list = ['/content/drive/MyDrive/Data Science /Tweet and NFT Sums Updated/AxieInfinityTweetsDataByMonth.csv', 
             '/content/drive/MyDrive/Data Science /Tweet and NFT Sums Updated/BoredApeKennelClubTweetsDataByMonth.csv',
             '/content/drive/MyDrive/Data Science /Tweet and NFT Sums Updated/BoredApeYachtClubTweetsDataByMonth.csv',
             '/content/drive/MyDrive/Data Science /Tweet and NFT Sums Updated/CoolCatsTweetsDataByMonth.csv',
             '/content/drive/MyDrive/Data Science /Tweet and NFT Sums Updated/CryptopunkTweetsDataByMonth.csv',
             '/content/drive/MyDrive/Data Science /Tweet and NFT Sums Updated/CyberKongsTweetsDataByMonth.csv',
             '/content/drive/MyDrive/Data Science /Tweet and NFT Sums Updated/FamersWorldTweetsDataByMonth.csv',
             '/content/drive/MyDrive/Data Science /Tweet and NFT Sums Updated/MutantApeYachtClubTweetsDataByMonth.csv',
             '/content/drive/MyDrive/Data Science /Tweet and NFT Sums Updated/ParallelAlphaTweetsDataByMonth.csv',
             '/content/drive/MyDrive/Data Science /Tweet and NFT Sums Updated/SorareTweetsDataByMonth.csv',
             '/content/drive/MyDrive/Data Science /Tweet and NFT Sums Updated/TheSandboxTweetsDataByMonth.csv',
             '/content/drive/MyDrive/Data Science /Tweet and NFT Sums Updated/VeeFriendsTweetsDataByMonth.csv',
             '/content/drive/MyDrive/Data Science /Tweet and NFT Sums Updated/WorldofWomenTweetsDataByMonth.csv',
             '/content/drive/MyDrive/Data Science /Tweet and NFT Sums Updated/artblocksTweetsDataByMonth.csv',
             '/content/drive/MyDrive/Data Science /Tweet and NFT Sums Updated/crypToadzTweetsDataByMonth.csv']

# iterate through each file and find the Pearson and Spearman correlation 
# coefficients
for i in range(len(file_list)):
  df = pd.read_csv(file_list[i])
  df_USD_pearson_corr = df.corrwith(df['Avg Sales(USD)'], method='pearson')
  df_USD_spearman_corr = df.corrwith(df['Avg Sales(USD)'], method='spearman')
  df_ETH_pearson_corr = df.corrwith(df['Avg Sales (ETH)'], method='pearson')
  df_ETH_spearman_corr = df.corrwith(df['Avg Sales (ETH)'], method='spearman')
  df_list = [df_USD_pearson_corr, df_USD_spearman_corr, df_ETH_pearson_corr, 
             df_ETH_spearman_corr]
  df_temp = pd.concat(df_list, axis=1, keys=["Pearson USD", "Spearman USD", 
                                             "Pearson ETH", "Spearman ETH"])
  df_final = df_temp.T
  # delete unwanted columns 
  del df_final['Total Transactions']
  del df_final['UniqueBuyers']
  del df_final['Sales(ETH)']
  del df_final['Avg Sales(USD)']
  del df_final['Avg Sales (ETH)']

  name = file_list[i].split("/")[6].partition('Tweets')[0]
  # Save file
  df_final.to_csv('/content/drive/MyDrive/Data Science /' + name + '_corr.csv')
```
