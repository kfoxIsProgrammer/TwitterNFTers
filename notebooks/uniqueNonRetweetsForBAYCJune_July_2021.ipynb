```python
import pandas as pd
import numpy as np
import io
import re



from google.colab import files
##IMport all CSV files
uploaded = files.upload()

##The data below is sensitive and cannot be made public
```

```python
fileTitles = ["BoredApeYachtClub_6_2021Tweets.csv", "BoredApeYachtClub_7_2021Tweets.csv"]

##Get the columns used in each csv for the newly made Dataframe
df = pd.read_csv(io.BytesIO(uploaded['BoredApeYachtClub_6_2021Tweets.csv']))
uniqueTweetsStrings = []
uniqueTweetFrame = pd.DataFrame(columns = df.columns)

for i in fileTitles:
  ##Get the dataframe for each csv
  dfVal = pd.read_csv(io.BytesIO(uploaded[i]))
  for i in range(len(dfVal.index)):
    ##Add the tweet if it is unique
    if(dfVal['text'][i] not in uniqueTweetsStrings):
      uniqueTweetFrame=uniqueTweetFrame.append([dfVal.iloc[i]])
      uniqueTweetsStrings.append(dfVal['text'][i])
      print(str(i) + "unique")
    else:
      print(str(i) + "Not unique")
```

```python
##Find unique Tweets that are not retweets
## IE retweets start with RT
fileTitles = ["BoredApeYachtClub_6_2021Tweets.csv", "BoredApeYachtClub_7_2021Tweets.csv"]

##Get the columns for the new Dataframe
df = pd.read_csv(io.BytesIO(uploaded['BoredApeYachtClub_6_2021Tweets.csv']))
uniqueTweetsStrings = []
uniqueTweetFrame = pd.DataFrame(columns = df.columns)

for i in fileTitles:
  dfVal = pd.read_csv(io.BytesIO(uploaded[i]))
  for i in range(len(dfVal.index)):
    ##Add the tweet if its unique and not a retweet
    if(dfVal['text'][i][:2] != "RT" and dfVal['text'][i] not in uniqueTweetsStrings):
      uniqueTweetFrame=uniqueTweetFrame.append([dfVal.iloc[i]])
      uniqueTweetsStrings.append(dfVal['text'][i])
      print(str(i) + "unique and not retweet")
    else:
      print(str(i) + "Not unique not retweets")
```
